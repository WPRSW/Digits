```python
import numpy as np 
import pandas as pd
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = load_digits()
df
```

```python
print(df["DESCR"])
```

```python
x = df["data"]
y = df["target"]
x.shape
```

```python
plt.imshow(np.reshape(x[1796,:],(8,8)))
```

```python
cdf = pd.DataFrame(x)
cdf["target"] = y
cdf
```

```python
train, test = train_test_split(cdf,test_size=0.2,random_state=42)
train
```

```python
train_target = train["target"]
train = train.drop("target",axis=1)
print(train.shape)
print(train_target.shape)
```

```python
test_target = test["target"]
test = test.drop("target",axis=1)
print(test.shape)
print(test.shape)
```

```python
train = train.values
train_target = train_target.values
test = test.values
test_target = test_target.values
```

# Make centers

```python
centers = []
for i in range(10):
    centers.append(train[train_target==i,:].mean(axis=0))
```

# Show

```python
import matplotlib.gridspec as gridspec
```

```python
fig = plt.figure(figsize=(15,8))
ga = gridspec.GridSpec(2,10)

for i in range(10):
    ax = fig.add_subplot(ga[0,i])
    plt.imshow(np.reshape(centers[i],(8,8)))
    plt.axis("off")
```

```python
range(test.shape[0])
```

```python
y_pred = []

for n in range(test.shape[0]):
    inf = np.inf
    for i in range(10):
        distance = np.linalg.norm(centers[i]-test[n,:],2)
        
        if distance < inf:
            inf = distance
            label = i
    y_pred.append(label)
y_pred = np.asarray(y_pred)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_true=test_target,y_pred=y_pred)
```

```python
sum(y_pred == test_target)/test.shape[0]
```
